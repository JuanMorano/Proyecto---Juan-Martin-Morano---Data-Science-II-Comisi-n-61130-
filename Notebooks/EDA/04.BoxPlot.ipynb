# ANALISIS DE BOXPLOT

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Función para clasificar columnas en continuas, discretas y categóricas
def classify_columns(df):
    # Identificar columnas numéricas y categóricas
    categorical_columns = df.select_dtypes(include=['object']).columns
    discrete_columns = [col for col in df.select_dtypes(include=['int']).columns if df[col].nunique() < 20]
    continuous_columns = [col for col in df.select_dtypes(include=['float', 'int']).columns if col not in discrete_columns]
    
    return discrete_columns, continuous_columns, categorical_columns

discrete_columns, continuous_columns, categorical_columns = classify_columns(df)

if 'Unnamed: 0' in continuous_columns:
    continuous_columns.remove('Unnamed: 0')

print("Discretas:", discrete_columns)
print("Continuas:", continuous_columns)
print("Categóricas:", categorical_columns)
```

## Visualizaciones Generales de Boxplot

```python
# Función para formatear valores según sean decimales o números grandes
def format_value(val):
    if abs(val) >= 1000:  # Si el valor es mayor que 1000, usar formato con comas
        return f"{val:,.0f}"
    elif isinstance(val, float):  # Si es decimal, usar hasta 2 decimales si es necesario
        return f"{val:.2f}"
    return str(val)

for col in continuous_columns:
    plt.figure(figsize=(8, 10))  # Ajusta la figura para visualización vertical
    ax = sns.boxplot(y=df[col])  # El gráfico será vertical (boxplot por defecto)
    plt.title(f'Boxplot de {col}')
    
    # Obtener estadísticas del boxplot
    stats = df[col].describe()
    IQR = stats['75%'] - stats['25%']  # Rango intercuartílico
    lower_bound = stats['25%'] - 1.5 * IQR
    upper_bound = stats['75%'] + 1.5 * IQR
    
    # Crear el cuadro de texto con los valores estadísticos formateados
    stats_text = (f"Min: {format_value(stats['min'])}\n"
                  f"Q1: {format_value(stats['25%'])}\n"
                  f"Mediana: {format_value(stats['50%'])}\n"
                  f"Q3: {format_value(stats['75%'])}\n"
                  f"Max: {format_value(stats['max'])}")

    # Posicionar el cuadro de texto dentro del gráfico
    plt.text(0.95, 0.85, stats_text, 
             horizontalalignment='right', verticalalignment='center', 
             transform=plt.gca().transAxes, 
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

    plt.show()
```
