# ANALISIS DE CORRELACION

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/jmorano/Documents/Plantillas personalizadas de Office/Coderhouse/Data Science/Modulo 2 - Data Science Fundamentos para la Ciencia de Datos II/Proyecto - Juan Martin Morano/Data/Processed/full_data_flightdelay_processed.csv', sep=',')
```

```python
# Función para clasificar columnas en continuas, discretas y categóricas
def classify_columns(df):
    # Identificar columnas numéricas y categóricas
    categorical_columns = df.select_dtypes(include=['object']).columns
    discrete_columns = [col for col in df.select_dtypes(include=['int']).columns if df[col].nunique() < 20]
    continuous_columns = [col for col in df.select_dtypes(include=['float', 'int']).columns if col not in discrete_columns]
    
    return discrete_columns, continuous_columns, categorical_columns

discrete_columns, continuous_columns, categorical_columns = classify_columns(df)

if 'Unnamed: 0' in continuous_columns:
    continuous_columns.remove('Unnamed: 0')

print("Discretas:", discrete_columns)
print("Continuas:", continuous_columns)
print("Categóricas:", categorical_columns)
```

## Correlaciones por Variables Continuas (Coeficiente de Pearson)

### Matrices Generales

```python
df_continuous = df[continuous_columns]
correlaciones_de_pearson = df_continuous.corr()
correlaciones_de_pearson
```

```python
# Generar el heatmap por coeficiente de Pearson
plt.figure(figsize=(10, 8))
sns.heatmap(correlaciones_de_pearson, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlaciones entre Variables Float')
plt.show()
```

### Matrices Aperturadas por Variable Target

```python
df_continuous['DEP_DEL15'] = df['DEP_DEL15']

# Filtrar los subconjuntos basados en el valor de 'DEP_DEL15'
df_target_1 = df_continuous[df_continuous['DEP_DEL15'] == 1]
df_target_0 = df_continuous[df_continuous['DEP_DEL15'] == 0]
df_target_1 = df_target_1.drop(columns=['DEP_DEL15'])
df_target_0 = df_target_0.drop(columns=['DEP_DEL15'])

# Calcular las correlaciones para cada subconjunto
correlation_target_1 = df_target_1.corr()
correlation_target_0 = df_target_0.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_target_1, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlaciones entre Variables Float')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_target_0, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlaciones entre Variables Float')
plt.show()
```

## Correlaciones por Variables Discretas (Coeficiente de Spearman)

### Matrices Generales

```python
df_discrete = df[discrete_columns]
correlaciones_de_spearman = df_discrete.corr(method='spearman')
correlaciones_de_spearman
```

```python
# Generar el heatmap por coeficiente de Spearman
plt.figure(figsize=(10, 8))
sns.heatmap(correlaciones_de_spearman, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlaciones entre Variables Float')
plt.show()
```

### Matrices Aperturadas por Variable Target

```python
# Filtrar los subconjuntos basados en el valor de 'DEP_DEL15'
df_target_d1 = df_discrete[df_discrete['DEP_DEL15'] == 1]
df_target_d0 = df_discrete[df_discrete['DEP_DEL15'] == 0]

# Eliminar la columna 'DEP_DEL15' en ambos subconjuntos para el cálculo de la correlación
df_target_d1 = df_target_d1.drop(columns=['DEP_DEL15'])
df_target_d0 = df_target_d0.drop(columns=['DEP_DEL15'])

# Calcular las correlaciones para cada subconjunto
correlaciones_de_spearman_1 = df_target_d1.corr()
correlaciones_de_spearman_0 = df_target_d0.corr()
```

```python
# Generar el heatmap por coeficiente de Spearman
plt.figure(figsize=(10, 8))
sns.heatmap(correlaciones_de_spearman_1, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlaciones entre Variables Float')
plt.show()
```

```python
# Generar el heatmap por coeficiente de Spearman
plt.figure(figsize=(10, 8))
sns.heatmap(correlaciones_de_spearman_0, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlaciones entre Variables Float')
plt.show()
```
