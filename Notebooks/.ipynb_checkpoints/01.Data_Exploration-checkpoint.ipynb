```python
# Sacamos el total de duplicados
total_duplicados = df.duplicated().sum()
print(f"Total de filas duplicadas: {total_duplicados}")

# Porcentaje de duplicados sobre el total de registros
porcentaje_duplicados = (total_duplicados / len(df)) * 100
print(f"Porcentaje de filas duplicadas: {porcentaje_duplicados:.2f}%")
```

```python
# Mapping (Diccionario para renombrar)
rename = {'mean': 'Promedio',
          '25%' : 'Q1',
          '50%' : 'Q2/Mediana',
          '75%' : 'Q3',
          'count' : 'Recuento',
          'std' : 'Desvio'}

df_desc = df.describe().T
df_desc = df_desc.rename(columns=rename)
df_desc = df_desc.round(2)
df_desc
```

```python
# Queremos analizar el peso del Desvio sobre el Promedio para anticipar la variabilidad en los datos
desvio_sobre_promedio = (df_desc['Desvio']/df_desc['Promedio'])*100
desvio_sobre_promedio
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        unique_values = df[column].unique()
        print(f'La columna: {column}:')
        print(f'Cantidad de categorias: {len(unique_values)}')
        print(f'valores unicos son: {unique_values}')
        print('-----------------------------------')
```

### Verificación del Desbalance de Clases de la Variable Target ('DEP_DEL15')

```python
df['DEP_DEL15'] = df['DEP_DEL15'].replace({1: 'Demoras', 0: 'Sin Demoras'})
# Contar la frecuencia de cada valor en la columna 'DEP_DEL15'
counts = df['DEP_DEL15'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(counts, labels=counts.index, startangle=140, 
        colors=sns.color_palette('muted'), wedgeprops={'edgecolor': 'black', 'linewidth': 2},
        textprops={'fontsize': 16}, autopct=lambda p: f'{p:.1f}%')
plt.title('Porcentaje de Desbalance de Clases de la Variable Target de Demora', fontsize=22)
plt.show()
```
